# Lab 03 - Analyze Logs
โหลดไฟล์ .csv ล่าสุดจาก `data/raw/` แล้ว plot ค่าเซนเซอร์แต่ละตัวแยก pane เทียบกับ time step (X-axis)

```python
import glob
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# หา path ของ project root (1 level ขึ้นจาก analysis/)
RAW_DATA_DIR = os.path.join('..', 'data', 'raw')

# ไล่หาไฟล์ .csv ทั้งหมดในทุก run แล้วเลือกไฟล์ที่แก้ไขล่าสุด
all_csv = glob.glob(os.path.join(RAW_DATA_DIR, '**', '*.csv'), recursive=True)
if not all_csv:
    raise FileNotFoundError(f'ไม่พบไฟล์ .csv ใน {RAW_DATA_DIR} — รัน main.py ก่อน')

latest_csv = max(all_csv, key=os.path.getmtime)
print('ใช้ไฟล์ log:', latest_csv)

df = pd.read_csv(latest_csv)
df.head()
```

```python
# ใช้ 'loop' (จำนวนครั้งที่ loop ทำงาน) เป็น time step บนแกน X
x = df['loop']

# คอลัมน์ที่จะ plot (ไม่รวม loop / timestamp / elapsed_time_s)
plot_columns = [c for c in df.columns if c not in ('loop', 'timestamp', 'elapsed_time_s')]

n = len(plot_columns)
fig, axes = plt.subplots(n, 1, figsize=(10, 2.2 * n), sharex=True)

if n == 1:
    axes = [axes]

for ax, col in zip(axes, plot_columns):
    ax.plot(x, df[col])
    ax.set_ylabel(col)
    ax.grid(True, alpha=0.3)

axes[-1].set_xlabel('Time step (loop)')
fig.suptitle('RoboMaster Sensor Log - Square Path (2x2 tiles)')
fig.tight_layout()

os.makedirs('plots', exist_ok=True)
fig.savefig(os.path.join('plots', 'sensor_log_plot.png'), dpi=150)
plt.show()
```
